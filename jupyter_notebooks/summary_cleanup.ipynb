```python
import pandas as pd

#load summary.tsv
df = pd.read_csv("summary.tsv", sep="\t")
df.head()
```

```python
#drop irrelevant columns
df = df.drop(columns=["aa_total", "proteins", "GC-content", "gravy_mean"])
df.head()
```

```python
#rename columns 
df = df.rename(columns={
    "temperature_optimum": "ogt",
    "tRNA_GC_mean": "trna_gc",
    "avg_tRNA_length": "trna_length",
    "GC3s": "gc3s",
    "Gravy": "gravy",
    "Aromo": "aromo",
    "Nc": "nc"
})

df.head()
```

```python
#convert genome size to Mbp
df["genome_size"] = df["genome_size"] / 1e6

#convert gc value to percent
df["gc3s"] = df["gc3s"] * 100

df.head()
```

```python
#round values

df = df.round({
    "ogt": 2,
    "genome_size": 2,
    "gene_density": 2,
    "frac_charged": 3,
    "frac_polar": 3,
    "frac_hydrophobic": 3,
    "frac_aromatic": 3,
    "frac_ilvwyg": 3,
    "trna_gc": 2,
    "trna_length": 2,
    "gc3s": 2,
    "gravy": 3,
    "aromo": 3,
    "nc": 3
})

df.head()
```

```python
#save file

df.to_csv("summary_clean.tsv", sep="\t", index=False) 
```
